and see its properties

####Problem 2:
Given a matrix and a vector, write a Python function that performs Matrix-Vector
product. Check your result with NumPy built-in function to perform Matrix-Vector/Matrix-Matrix product.

```python
import numpy as np

def dot_product(vec1, vec2):
    # flatten the input vectors
    vec1 = vec1.flatten()
    vec2 = vec2.flatten()
    return np.sum(np.multiply(vec1, vec2))

def matrix_vector_product(mat, vec):
    m, n = mat.shape
    p, q = vec.shape

    assert n == p, "The matrix and vector are not compatible for multiplication!"

    # create a ndarray to store the result
    result = np.zeros((m, q), dtype=float)

    # Loop to perform Matrix-Vector product
    for i in range(m):
        result[i] = dot_product(mat[i, :], vec)

    return result

# Example matrix and vector
mat1 = np.full((3, 3), 3, dtype=int)
vec1 = np.full((3, 1), 2, dtype=int)

print("Matrix =\n", mat1)
print("Vector =\n", vec1)

# Call the custom function
result1 = matrix_vector_product(mat1, vec1)
print("Result (custom function) =\n", result1)

# Check with numpy in-built function for multiplication
result2 = mat1 @ vec1
print("Result (NumPy @ operator) =\n", result2)
```
