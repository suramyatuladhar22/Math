#Part 1: Create and Understand Linear Systems:

###Example 1: We will start with a simple problem. Consider the following linear systems:

a) x+3y=5

𝑥−2𝑦=−5

𝑥+𝑦=3

b) 2x+3y=3

4𝑥+6𝑦=6

3𝑥−2𝑦=1

Solve the linear systems graphically.

```python
# a) Solution:
import numpy as np

import matplotlib.pyplot as plt

""" Part 1 - Problem 1: Solve linear systems graphically. """

# Generate 100 equidistant point between -5 to 5 for x

x = np.linspace(-5,5,100)

## Calculate y values for Equation 1 and 2

y1 = (5-x)/3

y2 = (5+x)/2

y3 = 3-x

## Generate a figure frame for plotting the lines

fig, ax = plt.subplots()

## Plot the two lines

ax.plot(x,y1, color='r')

ax.plot(x,y2, color='g')

ax.plot(x,y3, color='b')

## To decorate the plot, we insert equations.

ax.text(0.5,0.6, '$x+3y = 5$', color='r')

ax.text(-3,0.5,'$x-2y = -5$', color='g')

ax.text(1.2,2, '$x+y = 3$', color='b')

## Set the limits of the X & Y axes.

ax.set_xlim(-4,4)

ax.set_ylim(-2,6)

## Set property of X and Y axes, e.g., color and linewidth

ax.axvline(color='k',linewidth = 1)

ax.axhline(color='k',linewidth = 1)

## This option specifies the ticks based the list of numbers provided.

ax.set_xticks(list(range(-4,5)))

ax.set_aspect('equal')

ax.grid(True,ls=':')

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

```python
# 1) Solution:
print("(1)")
import numpy as np

# Create two random 5x6 matrices
A = np.random.randint(1, 10, (5, 6))
B = np.random.randint(1, 10, (5, 6))

# Add them together
C = A + B

# Print results
print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nResult (A + B):\n", C)

# 2) Solution:
print("\n(2)")
# Create three random 2x3 matrices
A = np.random.randint(1, 10, (2, 3))
B = np.random.randint(1, 10, (2, 3))
C = np.random.randint(1, 10, (2, 3))

# Perform (A - B) + C
Result = A - B + C

# Print results
print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nMatrix C:\n", C)
print("\nResult (A - B + C):\n", Result)

# 3) Solution:
print("\n(3)")
# Create a 4x4 matrix with random numbers between 1 and 20
M = np.random.randint(1, 21, (4, 4))
print("Matrix M:\n", M)

# Prompt user for input
num_str = input("Enter a number to search: ")

# Validate input
if num_str.isdigit():
    num = int(num_str)
    count = np.count_nonzero(M == num)
    if count > 0:
        print(f"The number {num} appears {count} times in the matrix.")
    else:
        print(f"The number {num} is not in the matrix.")
else:
    print("Invalid input. Please enter a valid integer.")
```

###Numpy.linalg

linalg is a submodule in NumPy, a popular Python library used for numerical computations, particularly in the domain of linear algebra. linalg stands for "linear algebra". This submodule provides functions and classes for various linear algebra operations, such as matrix decompositions, eigenvalue computations, matrix multiplication, solving linear equations, and many others.

Some useful linalg functions:

Copy and run the following codes and write what the code does?



1. linalg.det(a): calculates the determinant of the matrix

```python
import numpy as np

# Define a 2x2 matrix
a = np.array([[1, 2],[3, 4]])

# Compute determinant
det_a = np.linalg.det(a)

print(det_a)
```

2. linalg.matrix_rank(a): computes the rank of the matrix

```python
import numpy as np

# Define a 2x2 matrix
a = np.array([[1, 2],[3, 4]])

rank_a=np.linalg.matrix_rank(a)

print(rank_a)
```

3. linalg.inv(a): calculates the multiplicative inverse of a matrix

```python
import numpy as np

# Define a 2x2 matrix
a = np.array([[1, 2],[3, 4]])

a_inv=np.linalg.inv(a)

print(a_inv)
```

4. linalg.solve(a,b): solve a linear system of equations ax=b

 #given system

𝑥+2𝑦=1

3𝑥+5𝑦=2

```python
import numpy as np

# Coefficient matrix
a = np.array([[1, 2],[3, 5]])

# Constants (right-hand side)
b = np.array([1, 2])

# Solve the system
x = np.linalg.solve(a, b)

print(x)
```

5. linalg.pinv(a): compute the pseudo-inverse of a matrix.

```python
import numpy as np

a=np.random.randn(3,2)

b=np.linalg.pinv(a)

x=np.allclose(a,np.dot(a,np.dot(b,a)))
print(x)
```

###Task 3:

  1. Create a matrix 𝑨 of order 5 × 5 with random entries. Calculate the rank of the matrix 𝑨 to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?

  2. Generate a matrix 𝑨 of order 5 × 5 with random entries and generate a vector (column) 𝒙 of length 5 with random entries. Calculate Matrix-vector product 𝒃 = 𝑨𝒙. Now, we have a linear system, i.e. the matrix 𝑨 and a vector 𝒃. Solve this Linear System.

  3. Solve the following system.

      8x+3y−2z=9

      −4x+7y+5z==15

      3x+4y−12z=35

```python
import numpy as np

# 1) Rank and invertibility
print("1) Rank and Invertibility")
A1 = np.random.randn(5,5)
print("Matrix A1:\n", A1)

# Rank
rank = np.linalg.matrix_rank(A1)
print("\nRank of A1:", rank)

# Determinant (Other way)
det = np.linalg.det(A1)
print("Determinant of A1:", det)

# Invertibility check
if rank == 5 and det != 0:
    print("Matrix is invertible")
else:
    print("Matrix is not invertible")

# 2) Matrix-vector product and solving
print("\n2) Matrix-Vector Product and Solving")
A2 = np.random.randn(5,5)
x = np.random.randn(5)

b = np.dot(A2, x)
print("Matrix A2:\n", A2)
print("Vector x:\n", x)
print("Vector b = A2*x:\n", b)

# Solve system A2 x = b
x_solution = np.linalg.solve(A2, b)
print("Solved x:\n", x_solution)

# 3) Solve given system
print("\n3) Solve Given System")
A3 = np.array([[8, 3, -2],[-4, 7, 5],[3, 4, -12]])
b3 = np.array([9, 15, 35])

solution = np.linalg.solve(A3, b3)
print("Solution to given system [x, y, z]:", solution)

# Verify
lhs = np.dot(A3, solution)
print("Verification (A3 * solution):", lhs)
print("Right-hand side b3:", b3)
```
