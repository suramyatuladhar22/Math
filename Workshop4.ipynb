Task 2:

Following data shows the change in population of different animals from 2017 to 2022.

    a. Create a pie chart for the population of each year
    b. Can we show all the pie chart on the single pane? Write the code.
image.png

```python
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020, 2021, 2022]
bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 11, 6, 1]
whales = [80, 54, 100, 76, 93, 72]

labels = ["Bears", "Dolphins", "Whales"]

# Create subplots (2 rows, 3 columns for 6 years)
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()

# Plot pie chart for each year
for i, year in enumerate(years):
    # Directly use list indexing
    values = [bears[i], dolphins[i], whales[i]]
    axes[i].pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f"Year {year}")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
mycolours = ["red", "silver", "blue", "white", "green"]
myexplode = [0.3, 0, 0, 0, 0]

plt.pie(numbers, labels=colors, autopct='%1.1f%%', explode=myexplode, colors=mycolours)
plt.title("Car Colors Distribution (Exploded)")
plt.show()
```

B)

```python
import matplotlib.pyplot as plt

colors = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
mycolours = ["red", "silver", "blue", "white", "green"]

# Explode the first slice and add shadow
myexplode = [0.3, 0, 0, 0, 0]

plt.pie(numbers, labels=colors, autopct='%1.1f%%', explode=myexplode, shadow=True, colors=mycolours)
plt.title("Car Colors Distribution (Exploded + Shadow)")
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
mycolours = ["red", "silver", "blue", "white", "green"]

plt.pie(numbers, autopct='%1.1f%%', colors=mycolours)
plt.legend(title="Car Colours:", labels=colors)
plt.title("Car Colors Distribution with Legend")
plt.show()
```
