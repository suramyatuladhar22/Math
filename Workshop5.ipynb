| **Fast & convenient**   | Slower for large computations                                                | Much faster for mathematical/numerical tasks                                 |

Example 7:

Create two vectors in 2-dimensional (2D) space and plot them using Matplotlib.
Solution: A vector in 2D space can created as a 1D ndarray with two elements as:

    V = np.array([1,1])
    W = np.array([-2.0,-1])

To plot the above vectors, we will use the quiver() function from the pyplot module of Matplotlib library. The quiver() function is used to create 2D field or velocity plots and is perfect for our purpose of vector plotting. Here is an example of plotting the above vectors.

```python
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

Task: 6

a. Create three vectors in 3D space and plot them using Matplotlib.

    U= np.array([1, 0.25 ,0])
    V = np.array([0, 1, 0.25])
    W = np.array([0.25, 0, 1])

To do 3D plot, you will need to create 3D axes using following code:

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

```python
import numpy as np
import matplotlib.pyplot as plt

# Vectors
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

# Create 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors
ax.quiver(0,0,0, U[0],U[1],U[2], color='r', label='U')
ax.quiver(0,0,0, V[0],V[1],V[2], color='g', label='V')
ax.quiver(0,0,0, W[0],W[1],W[2], color='b', label='W')

ax.set_xlim([-1,2])
ax.set_ylim([-1,2])
ax.set_zlim([-1,2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.show()
```

b. Create two vectors in 2D space, and add them to create a third vector, and plot all three vectors using Matplotlib.

U = np.array([-0.5, 0.5])

V = np.array([1,1])

The two vectors are added as

W = U + V # adding two vector

```python
import numpy as np
import matplotlib.pyplot as plt

U = np.array([-0.5, 0.5])
V = np.array([1, 1])
W = U + V   # vector addition

fig, ax = plt.subplots()
ax.quiver(0,0, U[0],U[1], angles='xy', scale_units='xy', scale=1, color='r', label='U')
ax.quiver(0,0, V[0],V[1], angles='xy', scale_units='xy', scale=1, color='g', label='V')
ax.quiver(0,0, W[0],W[1], angles='xy', scale_units='xy', scale=1, color='b', label='W = U+V')

ax.set_xlim([-1,2])
ax.set_ylim([-1,2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
plt.grid()
plt.show()
```

c. Write a function to calculate magnitude of a vector. For example, calculate the magnitude of
the vector be U = (1.5, 1.5, 2.0).

```python
import numpy as np

def magnitude(vec):
    return np.linalg.norm(vec)

U = np.array([1.5, 1.5, 2.0])
print("Magnitude of U:", magnitude(U))
```

d. Write a function to calculate the direction cosines of the vector. For example, let the vector be

O = (1.5, 1.5, 2.0).

```python
import numpy as np

def direction_cosines(vec):
    mag = np.linalg.norm(vec)
    return vec / mag   # cosines with x, y, z axes

O = np.array([1.5, 1.5, 2.0])
print("Direction cosines of O:", direction_cosines(O))
```

e. Write a function to calculate Scalar (dot) product of two vectors. For example, let the vectors
be a= (1.5, 1.5, 2.0) and b = (−1.5, 0.5, 2.0).

```python
import numpy as np

def scalar_product(a, b):
    return np.dot(a, b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
print("Dot product:", scalar_product(a, b))
```

f. Write a function to calculate Vector (cross) product of two vectors. For example, let the vectors be

a = (1.5, 1.5, 2.0) and b= (−1.5, 0.5, 2.0)

```python
import numpy as np

def vector_product(a, b):
    return np.cross(a, b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
print("Cross product:", vector_product(a, b))
```
