1. Write a Python function that takes a list and returns a new list with distinct elements from the
first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```python
def newList():
  sample_list=[1,2,3,3,3,3,4,5]
  new_list=[]
  for i in range(len(sample_list)):
    if(sample_list[i] not in new_list):
      new_list.append(sample_list[i])
  print(new_list)
newList()
```

2. Write a Python function that takes a number as a parameter and checks whether the number
is prime or not.

```python
def prime(num):
  if(num<=1):
    print("not prime number")
    return

  for i in range(2,num):
    if(num%i==0):
      print("not prime number")
      return
  print("prime number")

number=int(input("Enter a number"))
prime(number)
```

3. Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
def prime():
  prime_list=[]
  count=0
  for n in range(3,1001,2):
      for i in range(2, n - 1):
          if (n % i == 0):
              count=count +1

      if (count == 0):
          prime_list.append(n)
      count=0

  for j in range(len(prime_list)-1):
      if prime_list[j+1] - prime_list[j] == 2:
          print(f"({prime_list[j]}, {prime_list[j+1]})")

prime()
```

4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sumOfDivisor(n):
  if(n<=1):
    print("the number is not perfect number")
    return

  sum=0
  for i in range(1,n):
    if(n%i==0):
      sum=sum+i
  if(n==sum):
    print("the number is perfect number")
  else:
    print("the number is not perfect number")

num=int(input("Enter a positive interger:"))
sumOfDivisor(num)
```

5. The smallest perfect number is 6. Write a program that finds and prints out the next perfect number. Hint:
You should be able your sumOfDiviors function from your solution to previous question.

```python
def sumOfDivisor(n):
  if n <= 1:
    return 0

  sum = 0
  for i in range(1, n):
    if n % i == 0:
      sum=sum+i
  return sum

def isPerfect(n):
  return n == sumOfDivisor(n)

num=7
number=num-1
while True:
    if isPerfect(num):
        print("The next perfect number after",number,"is:", num)
        break
    num += 1
```

6. Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
def mean(marks):
  total=0
  for i in marks:
    total=total+i
  mean=total/len(marks)
  print("mean =",mean)

def median(marks):
  marks.sort()
  n=len(marks)
  mid=n//2

  if(n%2==0):   #even length
    median=marks[mid-1]+marks[mid]/2
    print("median=",median)
  else:         #odd length
    median=marks[mid]
    print("median=",median)

Marks=[20,30,35,55,70,80,90]
mean(Marks)
median(Marks)
```

7. Write a function that converts a decimal number to binary number.

```python
def binary(dec):
  bin=[]
  while(dec>0):
    r=dec%2
    dec=dec//2
    bin.append(str(r))
  new_bin=''.join(bin[::-1])
  print(new_bin)


decimal=int(input("enter a decimal base 10 number: "))
binary(decimal)
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(num):
  product = 1
  while num > 0:
    digit = num % 10
    product *= digit
    num = num // 10
  return product

number = int(input("Enter a number: "))
print("Product of digits:", prodDigits(number))
```
